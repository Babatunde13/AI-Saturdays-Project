```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split, KFold
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

#import metrics functions
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
#Split data

X_train, X_test, y_train, y_test = train_test_split(
    data.drop('quality', axis=1), 
    data.quality, 
    test_size=0.3, 
    random_state=42)

X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
from sklearn.metrics import precision_recall_curve, precision_score, recall_score, confusion_matrix


pipe = Pipeline([
    ('transform', StandardScaler()),
    ('poly', PolynomialFeatures(degree=2)),
    ('clf', LogisticRegression(solver='newton-cg'))
])

pipe.fit(X_train, y_train)
print(pipe.score(X_test, y_test), pipe.score(X_train, y_train.to_frame()))
print(recall_score(pipe.predict(X_test), y_test, average='weighted'), recall_score(pipe.predict(X_train), y_train, average='weighted'))
precision_score(pipe.predict(X_test), y_test, average='weighted'), precision_score(pipe.predict(X_train), y_train, average='weighted')
```

```python
param_grid={
    'clf__penalty':['l1', 'l2'],
    'clf__tol': [11e-4, 1e-3, 0.0334],
    'clf__C': [0.01, 0.02, 0.3, 0.7, 0.8, 1.0, 1.2],
    'clf__random_state': [1, 42],
    'clf__solver': ['newton_cg', ]
}
```

```python
DecisionTreeClassifier
pipe = Pipeline([
    ('transform', StandardScaler()),
    ('poly', PolynomialFeatures(degree=2)),
    ('clf', RandomForestClassifier(n_estimators=1000))
])

pipe.fit(X_train, y_train)
print(pipe.score(X_test, y_test), pipe.score(X_train, y_train))
print(recall_score(pipe.predict(X_test), y_test, average='weighted'), recall_score(pipe.predict(X_train), y_train, average='weighted'))
precision_score(pipe.predict(X_test), y_test, average='weighted'), precision_score(pipe.predict(X_train), y_train, average='weighted')
```

```python

pipe = Pipeline([
    ('transform', StandardScaler()),
    ('poly', PolynomialFeatures(degree=2)),
#     ('pca', PCA(n_components=30)),
    ('clf', GradientBoostingClassifier(learning_rate=0.0353, n_estimators=200, criterion='mse'))
#     ('clf', RandomForestClassifier())
])

pipe.fit(X_train, y_train)
print(pipe.score(X_test, y_test), pipe.score(X_train, y_train.to_frame()))
print(recall_score(pipe.predict(X_test), y_test, average='weighted'), recall_score(pipe.predict(X_train), y_train, average='weighted'))
precision_score(pipe.predict(X_test), y_test, average='weighted'), precision_score(pipe.predict(X_train), y_train, average='weighted')
```

```python

pipe = Pipeline([
    ('transform', StandardScaler()),
    ('poly', PolynomialFeatures(degree=2)),
    ('clf', RandomForestClassifier(n_estimators=1000))
])

pipe.fit(X_train, y_train)
print(pipe.score(X_test, y_test), pipe.score(X_train, y_train))
print(recall_score(pipe.predict(X_test), y_test, average='weighted'), recall_score(pipe.predict(X_train), y_train, average='weighted'))
precision_score(pipe.predict(X_test), y_test, average='weighted'), precision_score(pipe.predict(X_train), y_train, average='weighted')
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV

param_grid = {
    'poly__degree':range(1, 5),
    'poly__include_bias':[True, False],
    'poly__interaction_only': [True, False],
    'clf__n_estimators':range(10, 110, 10),
    'clf__criterion': ['gini', 'entropy'],
    'clf__max_features': range(1, len(X_train.columns)),
    'clf__bootstrap': [True, False],
    'clf__random_state': range(1, 5)
}

model = RandomizedSearchCV(pipe, param_distributions=param_grid, cv=5, verbose=5, random_state=42, return_train_score=True)
model.fit(X_train, y_train)
```

```python
print(model.best_score_, model.best_params_)
print(model.score(X_test, y_test), model.score(X_train, y_train))
print(recall_score(pipe.predict(X_test), y_test, average='weighted'), recall_score(pipe.predict(X_train), y_train, average='weighted'))
precision_score(pipe.predict(X_test), y_test, average='weighted'), precision_score(pipe.predict(X_train), y_train, average='weighted')
```
